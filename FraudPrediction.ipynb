```python
import os
import pyttsx3
import numpy as np

# Initialize the text-to-speech engine
engine = pyttsx3.init()  # Initialize the pyttsx3

# Function to convert text to speech
def text_to_speech(text):
    engine.say(text)
    engine.runAndWait()  # Ensure the speech is processed before moving on to the next iteration
    engine.setProperty('rate', 150)  # Speed of speech
    engine.setProperty('volume', 1)  # Volume level (0.0 to 1.0)

# Function to get a new transaction
def get_new_transaction(index):
    return X_test_array[index, :]

index_file = "index.txt"

# Check if index.txt exists, and if not, create it and initialize with 0
if not os.path.exists(index_file):
    with open(index_file, "w") as f:
        f.write("0")

# Read the current index from the file
with open(index_file, "r") as f:
    index = int(f.read().strip())

# Check if index is out of bounds and reset to 0 if necessary
if index >= len(X_test_array):
    print("Index out of bounds. Resetting index to 0.")
    index = 0


new_transaction = get_new_transaction(index)

# Convert the new transaction into a DataFrame with correct feature names
new_transaction_df = np.array(new_transaction).reshape(1, -1)


# Make the prediction
prediction = loaded_model.predict(new_transaction_df)

# Determine if it's Fraud or Non-Fraud
if prediction[0] == 1:
    result = "Fraud"
else:
    result = "Non-Fraud"

# Print each row of X_test with its corresponding prediction
for i in range(len(new_transaction_df)):
    print(f"Predicted: {result}")


# Speak the result of the prediction
text_to_speech(f"The predicted result is {result}")

#print(X_test_array[index])

# Increment the index for the next run
index += 1

# Save the updated index back to the file
with open(index_file, "w") as f:
    f.write(str(index))
```
